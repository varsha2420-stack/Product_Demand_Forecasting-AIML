```python
# ==========================================================
# Retail Demand Forecasting
# Notebook - 2 : Model Building
# ==========================================================

# Used for reading CSV files
import pandas as pd

# Used for numerical calculations
import numpy as np

# Used for splitting dataset into training and testing data
from sklearn.model_selection import train_test_split

# Random Forest Regression Algorithm
from sklearn.ensemble import RandomForestRegressor

# XGBoost Regression Algorithm
from xgboost import XGBRegressor

# Used to evaluate model performance
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

# Used to save trained models
import joblib
```

```python
# ==========================================================
# Load Cleaned Dataset
# ==========================================================

# Read cleaned dataset

data = pd.read_csv("retail_sales_cleaned.csv")

# Display first five rows

data.head()
```

```python
# ==========================================================
# Dataset Shape
# ==========================================================

print("Number of Rows :", data.shape[0])

print("Number of Columns :", data.shape[1])
```

```python
# ==========================================================
# Input Features and Target Variable
# ==========================================================

# Input Features

X = data.drop(
    ["sales", "date"],
    axis=1
)

# Target Variable

y = data["sales"]

# Display first five input records

X.head()
```

```python
# ==========================================================
# Split Dataset into Training and Testing Data
# ==========================================================

X_train, X_test, y_train, y_test = train_test_split(

    X,

    y,

    test_size=0.20,

    random_state=42

)

print("Training Shape :", X_train.shape)

print("Testing Shape :", X_test.shape)
```

```python
# ==========================================================
# Reduce Training Data Size
# ==========================================================

# Take a random sample of 100000 training records

X_train = X_train.sample(n=100000, random_state=42)

# Select corresponding target values
y_train = y_train.loc[X_train.index]

print("New Training Shape :", X_train.shape)
print("New Target Shape   :", y_train.shape)
```

```python
# ==========================================================
# Random Forest Regressor
# ==========================================================

rf_model = RandomForestRegressor(

    n_estimators=50,

    max_depth=15,

    random_state=42,

    n_jobs=-1

)

# Train the model

rf_model.fit(

    X_train,

    y_train

)

print("Random Forest Model Trained Successfully")
```

```python
# ==========================================================
# Random Forest Prediction
# ==========================================================

rf_predictions = rf_model.predict(X_test)

print("Prediction Completed")
```

```python
# ==========================================================
# Random Forest Evaluation
# ==========================================================

rf_mae = mean_absolute_error(y_test, rf_predictions)
rf_rmse = np.sqrt(mean_squared_error(y_test, rf_predictions))
rf_r2 = r2_score(y_test, rf_predictions)

print("Random Forest Performance")
print("-------------------------")
print("MAE  :", rf_mae)
print("RMSE :", rf_rmse)
print("R2 Score :", rf_r2)
```

```python
# ==========================================================
# XGBoost Regressor
# ==========================================================

xgb_model = XGBRegressor(

    n_estimators=100,

    learning_rate=0.1,

    max_depth=6,

    random_state=42,

    n_jobs=-1

)

# Train the model

xgb_model.fit(

    X_train,

    y_train

)

print("XGBoost Model Trained Successfully")
```

```python
# ==========================================================
# XGBoost Prediction
# ==========================================================

xgb_predictions = xgb_model.predict(X_test)

print("Prediction Completed")
```

```python
# ==========================================================
# XGBoost Evaluation
# ==========================================================

xgb_mae = mean_absolute_error(y_test, xgb_predictions)
xgb_rmse = np.sqrt(mean_squared_error(y_test, xgb_predictions))
xgb_r2 = r2_score(y_test, xgb_predictions)

print("XGBoost Performance")
print("-------------------------")
print("MAE      :", xgb_mae)
print("RMSE     :", xgb_rmse)
print("R2 Score :", xgb_r2)
```

```python
# ==========================================================
# Save XGBoost Model
# ==========================================================

joblib.dump(xgb_model, "xgboost_model.pkl")

print("XGBoost Model Saved Successfully")
```

```python
comparison = pd.DataFrame({

    "Model": [
        "Random Forest",
        "XGBoost"
    ],

    "MAE": [
        rf_mae,
        xgb_mae
    ],

    "RMSE": [
        rf_rmse,
        xgb_rmse
    ],

    "R2 Score": [
        rf_r2,
        xgb_r2
    ]

})

comparison
```

```python
# ==========================================================
# Best Model
# ==========================================================

comparison = comparison.sort_values(
    by="R2 Score",
    ascending=False
)

comparison
```

```python
comparison.to_csv(
    "model_comparison.csv",
    index=False
)

print("Model Comparison Saved Successfully")
```
